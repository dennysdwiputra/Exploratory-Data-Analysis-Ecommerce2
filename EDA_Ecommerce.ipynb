## Define Problem

Dalam bisnis ecommerce, harga dan diskon sangat mempengaruhi penjualan suatu barang, bedasarkan data diatas, saya mendifine bahwa:
1. Apakah Free Shipping dapat mempengaruhi persentase penjualan dari suatu barang?
2. Apakah harga diskon yang tinggi (bedasarkan persentase diskon) dapat mempengarui penjualan?

## Tujuan EDA
1. memberikan promo free shiping pada setiap barang dan melakukan kerja sama ke pihak shipping untuk meningkatkan profit kedua belah pihak
2. Memberikan trik diskon besar besaran pada suatu periode untuk kategori dengan penjualan terbanyak

Terlebih dahulu kita memisahkan kategori antara Free shipping dan Eligible for Free shipping dimana Eligible for Free Shipping ini adalah Free shipping yang mempunyai syarat dan ketentuan tertentu untuk mendapatkannya

```python
freeship = ecomm[ecomm['shpping_rate']=='FREE Shipping']
```

## membuat kolom baru 

kolom baru ini menunjukkan persentase diskon yang belom di konversi, masi berupa 0,xx bukan 1% atau 50%

```python
freeship['persentase diskon']=((freeship['Original_price'])-(freeship['price']))/freeship['Original_price']
```

```python
elship['persentase diskon']=((elship['Original_price'])-(elship['price']))/elship['Original_price']
```
